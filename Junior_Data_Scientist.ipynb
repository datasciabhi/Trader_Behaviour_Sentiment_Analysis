Ans: Significant shifts in trader behavior are observed as sentiment changes. The analysis of the dataset is detailed in the four points below:

1. Trade Frequency (Biggest Change):
    
(a) Extreme Fear: Trading activity is highest here. Traders execute excessive trades in panic or to recover losses (over-trading).

Key Insight: Traders exhibit 'Revenge Trading' behavior during Extreme Fear (High frequency, Low profit)  and 'Strategic Trading' during Extreme Greed (Lower frequency, High profit)."

```python
# 3. Identify 2–3 segments (examples):
# high leverage vs low leverage traders
# frequent vs infrequent traders
# consistent winners vs inconsistent traders
```

```python
# High Leverage  vs Low Leverage Traders
#a. Summarizing behaviour of each account

trader_behaviour= df_merged.groupby('Account').agg({'Size USD':'mean','date':'count','Closed PnL':'sum'}).rename(columns={'date':'Trade_Frequency','Size USD':'Avg_Trade_Size'})
trader_behaviour.head()
```

```python
#b. After Calculating Median Making Segments
freq_median= trader_behaviour['Trade_Frequency'].median()
#print(freq_median)
#Classifying to Traders
trader_behaviour['Segment']=trader_behaviour['Trade_Frequency'].apply(lambda x:'High Frequency' if x>freq_median else 'Low Frequency')
print("Treader Behaviour Created Successfully: ")
print(trader_behaviour['Segment'].value_counts())
```

```python
# Merging Segments with the Main DataFrames
df_final= df_merged.merge(trader_behaviour[['Segment']],on='Account')
df_final.head()
```

```python
# Checking the avg Profit for each segment in Fear vs Greed
segment_performance= df_final.groupby(['Segment','classification'])['Closed PnL'].mean().unstack()	
print("Sentiment Wise Performance For Segments: ")
segment_performance
```

```python
# Chart 3: Segment-wise Profitability
plt.figure(figsize=(12, 6))
sns.boxplot(x='Segment', y='Closed PnL', hue='classification', data= df_final)
plt.ylim(-500, 1000) # Outliers manage karne ke liye
plt.title('Profit Distribution: High Frequency vs Low Frequency Traders')
plt.show()
```

Data analysis confirms that inconsistent traders suffer the most losses due to the high number of outliers on 'Fear' (Green) days. 
On the other hand, consistent winners are those who adjust their frequency according to market sentiment 
(lower in Fear and trend following in Greed)

a. Strategy: When market sentiment is in the 'Extreme Fear' zone, the system should reduce position size by 50% and limit leverage.

Rule 2: The "Contrarian Greed" Rule (Execution Strategy) -
    
a. Strategy: When sentiment hits 'Extreme Greed', the system should prioritize short-selling (mean-reversion) setups rather than blindly chasing longs.
    
b. Data-Driven Reasoning: Although the average profit in 'Extreme Greed' is the highest (205.81 dollar), the data also shows that 
    the Long/Short Ratio has dropped to 0.41. This means that smart money is closing positions and the market is ready for a reversal. 
    Contrarian trades (shorting) have a better risk-to-reward ratio during this phase.
